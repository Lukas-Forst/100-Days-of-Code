```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import os
```

```python
os.listdir('./archive')
```

```python
files = ["./archive/"+i for i in os.listdir('./archive')]
files
```

```python
df = pd.concat(
    map(pd.read_csv, files), ignore_index=True)
print(df)
```

```python
df.dtypes
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.dtypes
```

```python
df.head()
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
sns.barplot(x="Symbol", y="Marketcap", palette="rocket", data=df)
```

```python
sns.lineplot(data=df, x="Date", y="Marketcap", hue="Symbol")
```

```python
df_2017 = df[df['Date'] > "01-01-2017"]
```

```python
sns.lineplot(data=df_2017, x="Date", y="Marketcap", hue="Symbol")
```
